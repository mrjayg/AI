```python
import matplotlib.pyplot as plt
import time

def selection_sort(arr):
    """
    Function to perform selection sort and visualize each step.
    The selection sort algorithm selects the minimum element from the unsorted
    part and swaps it with the leftmost unsorted element.
    """
    n = len(arr)
    for i in range(n):
        min_idx = i
        for j in range(i + 1, n):
            if arr[j] < arr[min_idx]:
                min_idx = j
        # Swap the found minimum element with the first element
        arr[i], arr[min_idx] = arr[min_idx], arr[i]
        # Visualize after each swap
        visualize_selection_sort(arr, i, min_idx)
        plt.pause(0.5)  # Pause for visualization
    return arr

def visualize_selection_sort(arr, i, min_idx):
    """
    Visualize the array after each step of selection sort.
    This function plots a bar graph with different colors for the selected
    element and the minimum element.
    """
    plt.clf()  # Clear previous figure
    plt.bar(range(len(arr)), arr, color='lightblue')  # Default bars
    plt.bar(i, arr[i], color='green')  # Current element being swapped
    plt.bar(min_idx, arr[min_idx], color='red')  # Minimum element found
    plt.title(f"Selection Sort - Iteration {i + 1}")
    plt.draw()  # Draw updated figure

# Input array from the user
arr = list(map(int, input("Enter the array to sort (space separated): ").split()))

# Enable interactive mode for Matplotlib
plt.ion()

# Run Selection Sort with visualization
sorted_arr = selection_sort(arr)

# Display final sorted array
plt.clf()
plt.bar(range(len(arr)), arr, color='lightgreen')  # Final sorted array in green
plt.title("Final Sorted Array")
plt.show()

print(f"Sorted Array: {sorted_arr}")
```
